### Exercício 1
Utilizando o built-in method input(), crie um programa que receba a altura e o peso de uma pessoa e imprima na tela o IMC da mesma.

```python
import time

print('='*10 + 'Calculadora IMC' + '='*10)

height = float(input("\nAltura: "))
weight = float(input("Peso: "))

imc = weight / height**2

print('\nCalculando...\n')
time.sleep(1)
print('='*35)

print(f'\nIMC: {imc:.2f}')
```

### Exercício 4 
Faça um programa para uma loja de tintas. A pessoa informa a área em m2 que deseja pintar, e o script calculará a quantidade de latas de tinta que a pessoa deve comprar e o valor. Considere que cada litro de tinta pinta 3m2, que cada lata contém 18L e que custa R$ 80.

```python
#Cada litro de tinta pinta 3m2
#Cada lata contém 18L e que custa R$ 80.
#Cada lata de tinta pinta no máximo 54m²

area = float(input('Área(m²) que deseja pintar: '))

litros = area / 3
qtd_latas = litros / 18
valor_total = qtd_latas * 80

print(f'Quantidade Latas Necessárias: {qtd_latas:.1f} latas\nValor: R$ {valor_total:,.2f}')
```

### Exercício 5
Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o Imposto de Renda, 8% para o INSS e 5% para o sindicato, faça um programa que nos dê:
1. Salário bruto.
2. Quanto pagou ao INSS.
3. Quanto pagou ao sindicato.
4. O salário líquido.

```python
from IPython.core.display import clear_output

desconto_ir = 11/100
desconto_inss = 8/100
desconto_sindicato = 5/100

salario_hora = float(input('Salário por Hora: '))
horas_trabalhadas = float(input('Horas trabalhadas no mês: '))

#Salario Total
total_salario = salario_hora * horas_trabalhadas

#Descontos
valor_ir = total_salario * desconto_ir #Valor pago ao Imposto de Renda
valor_inss = (total_salario - valor_ir) * desconto_inss #Valor pago ao INSS
valor_sindicato = (total_salario - valor_ir - valor_inss)* desconto_sindicato #Valor pago ao Sindicato

#Salario Líquido
salario_liquido = total_salario - (valor_inss + valor_sindicato + valor_ir)

clear_output()

print(f'Salário:\n\nSalário Total: R$ {total_salario:,.2f}\nSalário Líquido: R$ {salario_liquido:,.2f}\n\nDescontos:\n\nValor pago ao Imposto de Renda: R$ {valor_ir:,.2f}\nValor pago ao INSS: R$ {valor_inss:,.2f}\nValor pago ao Sindicato: R$ {valor_sindicato:,.2f}\n')
```
